```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = ["Arial"]
plt.rcParams["axes.unicode_minus"] = False

df = pd.read_csv("gdp_growth.csv", skiprows=4)

countries = ["China", "United States", "Germany", "Japan", "United Kingdom"]
df = df[df["Country Name"].isin(countries)]

df = df[["Country Name", "2020", "2021", "2022", "2023", "2024"]]

print("Filtered GDP Growth Data (2020-2024):")
print(df)
```

```python
df = df.dropna()
print("\nData after cleaning:")
print(df)
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
plt.figure(figsize=(12, 6))

for index, row in df.iterrows():
    years = ["2020", "2021", "2022", "2023", "2024"]
    values = [row["2020"], row["2021"], row["2022"], row["2023"], row["2024"]]
    plt.plot(years, values, marker='o', label=row["Country Name"])

plt.title("GDP Growth Rate Comparison: Major Economies (2020-2024)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("GDP Growth Rate (%)")
plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)
plt.show()
```
